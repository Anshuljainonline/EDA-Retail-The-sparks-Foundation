## Looking for Outliers 

```python
#looking for outliers in the "profit" column
sns.boxplot(x=df.Profit)
```

```python
#looking for outliers in the "discount" column
sns.boxplot(x=df.Discount)
```

### Removing the outliers

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR= Q3-Q1 #interquartile Range
print(IQR)
```

```python
df2=df[~((df < (Q1-1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]
df2.shape
```

```python
#confiring removal of outliers in the "profit" column
sns.boxplot(x=df2.Profit)
```

```python
#confiring removal of outliers in the "Discount" column
sns.boxplot(x=df2.Discount)
```

## Profits division on the basis of Segment

```python
d1={}
for segment in df2.Segment.unique():
    mean_profit_state= df[df.Segment== segment].Profit.mean()
    print(mean_profit_state)
    d1[segment]=mean_profit_state
```

```python
fig= plt.figure(figsize=(12,8))
plt.bar(d1.keys(), d1.values())
plt.show

#implies that maximum total profits come fromHome office products but the difference between it and corporate is not much
```

## Different profits on basis of State

```python
df2.State.unique()
```

```python
sns.countplot(df2.State)
```

```python
d2={}
for state in df2.State.unique():
    mean_state_profit= df2[df2.State==state].Profit.sum()/len(df2[df2.State==state])
    d2[state]= mean_state_profit
    print(mean_state_profit)

sorted_tuples = sorted(d2.items(), key=lambda item: item[1])
```

```python
sorted_tuples
```

```python
fig=plt.figure(figsize=(32,20))
plt.bar(d2.keys(),d2.values())
plt.show()

#signifies that the profits are high in
```
